```python
import keras_ocr as kr
import matplotlib.pyplot as plt
```

```python
trained_data = kr.pipeline.Pipeline()
```

```python
ocr_images = [
    kr.tools.read(url) for url in [
      'https://i.postimg.cc/BZ7JfvzV/00043445-00043449.png',
      'https://i.postimg.cc/g0mCZ95h/0000223278.png',
      'https://i.postimg.cc/25HCj86y/00555341.png'
    ]
]
```

```python
plt.figure(figsize=(10,10))
plt.imshow(ocr_images[0])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(ocr_images[1])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(ocr_images[2])
```

```python
recognizing_models = trained_data.recognize(ocr_images)
```

```python
fig, axs = plt.subplots(nrows=len(ocr_images), figsize=(15, 15))
for ax, image, predictions in zip(axs, ocr_images, recognizing_models):
    kr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
print(predictions)
```
